```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.head(10)
#nba.info()
```

```python
nba[["Name","Weight"]].head(5)
#will get data from these two columns
```

```python
nba["Sport"] = "Football"
nba.head(10)
#will create an additional column called Sport and will assign the value football to all the rows
```

```python
nba.insert(3, column = "Sport1", value = "Football")
nba.head(10)
#will inset the column at index posn 3
```

```python
nba.dropna(subset = ["Salary", "College"])
#will drop the defined rows
```

```python
nba.fillna(0).head(30)
#will fill the NaN or null values wit 0
```

```python
nba["Salary"] = nba["Salary"].astype("int")
#will convert the values into integer
```

```python
nba[nba["Team"] == "Boston Celtics"]
#will get the team name called Boston Celtics from all the rows in column team
```

```python
bond = pd.read_csv("jamesbond.csv")
bond.head(10)
```

```python
bond.set_index("Film", inplace = True)
```

```python
bond.reset_index(inplace = True)
```

```python
bond.drop(["Actor", "Director"], axis = 1)
```

```python
#uses index value to find the remaining values
bond.loc["Goldfinger"]
#the first parameter passed is related to select the row
```

```python
bond.loc["Casino Royale", ["Actor", "Box Office"]]
#the first parameter Casion Royale passed indicates Row, and the second list passsed indicates the columns
```

```python
bond.iloc[10, 2:5]
#selects the values based on their index no.
```
